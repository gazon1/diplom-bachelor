```python
def get_url(key, coordinates, time=None):
    """
    coordinates - list of 2 decimals: lattitude and longitude
    key - API secret key
    
    returns: url to pass to the requests.get method 
    """
    url = f"https://api.darksky.net/forecast/{key}/{coordinates[1]},{coordinates[2]}"
    if time:
        url += f",{time}"
    return url
```
